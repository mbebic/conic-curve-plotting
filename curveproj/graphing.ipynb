```python
from plotly.offline import download_plotlyjs, plot
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
#general canonical form for ellipse with xc = 1, yc = 1, a =2, b=1
a = 0.25
b = 0
c = 1
d = -0.5
e = -2
f = 0.25
```

```python
mx = np.array([[a,b/2],[b/2,c]])
print(mx)
```

```python
[lam1,lam2] = np.linalg.eig(mx)[0]
print(lam1, lam2)
```

```python
mx2 = np.array([[a,b/2,d/2],[b/2,c,e/2],[d/2,e/2,f]])
print(mx2)
```

```python
s = np.linalg.det(mx2)
print(s)
```

```python
a = np.sqrt(-s/(lam1**2*lam2))
b = np.sqrt(-s/(lam1*lam2**2))
print(a,b)
```

```python
xc = -d*a**2/2
yc = -e*b**2/2
print(xc,yc)
```

```python
theta = np.linspace(0,2*np.pi,200)
x = xc+a*np.cos(theta)
y = yc+b*np.sin(theta)
```

```python
fig = px.line(x=x,y=y)
fig.show()
```

```python
fig.update_layout(autosize=False)
```

```python
graph = fig.to_html(full_html=False, default_height=600, default_width=600)
f = open('graph.html', "w")
f.write(graph)
f.close()
```
